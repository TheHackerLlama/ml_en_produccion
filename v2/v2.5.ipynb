```
import os

train_path = 'train'
test_path = 'test'
```

```
import numpy as np
from tensorflow.python.keras.preprocessing.image import load_img, img_to_array
from tensorflow.python import keras
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator

img_size = 224
num_classes = 2

data_generator = ImageDataGenerator()

train_generator = data_generator.flow_from_directory(
        train_path,
        target_size=(img_size, img_size),
        batch_size=498,
        class_mode='categorical')

test_generator = data_generator.flow_from_directory(
        test_path,
        target_size=(img_size, img_size),
        batch_size=500,
        class_mode='categorical')
```

```
print(len(train_generator))
print(len(train_generator[0]))
print(train_generator[0][0].shape)
```

```
input_x = (train_generator[0][0]/255)
input_y = (train_generator[0][1])
```

```
output_x = (test_generator[0][0]/255)
output_y = test_generator[0][1]
```

```
input_y = np.delete(input_y, 1, 1)
output_y = np.delete(test_generator[0][1], 1, 1)
```

```
import autokeras as ak

clf = ak.ImageClassifier(verbose=True)
clf.fit(input_x, input_y, time_limit=2*60)
clf.final_fit(input_x, input_y, output_x, output_y, retrain=True)
clf.export_autokeras_model('my_model.h5')
```

```
model = ak.utils.pickle_from_file('my_model.h5')
results = model.evaluate(output_x, output_y)
print(results)
```
